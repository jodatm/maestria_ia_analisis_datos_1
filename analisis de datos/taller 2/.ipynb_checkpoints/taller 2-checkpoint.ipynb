```python
# Librerias
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ruta de archivo 
ruta_archivo = 'ObesityDataSet_raw_and_data_sinthetic.csv'

data = pd.read_csv(ruta_archivo)
```

```python
# Imprimir informaci칩n b치sica sobre el conjunto de datos

print("Informaci칩n del conjunto de datos:")
print(data.info())
```

```python
# Primeras filas del conjunto de datos
data.head(5)
```

```python
# Estad칤sticas descriptivas
print(data.describe())
```

```python
# El dataset esta limpio, no hay datos nulos. Se decide seleccionar el atributo FAVC (Do you eat high caloric food frequently?)
# Este atributo tiene datos entre 1 y 3, donde 1 es consume poco este tipo de alimentos y 3 es mucho.
# Existen estudios que respaldan la idea del consumo de alimentos altos en calorias con la obesidad
# https://www.nhlbi.nih.gov/es/salud/sobrepeso-y-obesidad/causas
# https://www.scielo.org.mx/scielo.php?script=sci_arttext&pid=S1665-11462006000300006
```

```python
#Econtremos c칩mo est치n repartidas las categor칤as en el conjunto de datos.
conteo_de_ejemplos_por_clase = data['FAVC'].value_counts()

#Revisemos esta informaci칩n de forma gr치fica.
#-------------------------------------------------------------------------------
#Usemos un gr치fico tipo pastel desde pandas.
#-------------------------------------------------------------------------------
conteo_de_ejemplos_por_clase.plot(kind='pie',
                                    figsize=(5, 5),
                                    autopct=lambda pct: '{:.1f}%\n({:d})'.format(pct, int(pct/100.*conteo_de_ejemplos_por_clase.sum())), #Con esta funci칩n lambda se puede mostrar el porcentaje y el n칰mero absoluto de ejemplos en cada categor칤a.
                                    startangle=90,       #Esto define la orientaci칩n de la l칤nea vertical inicial de las divisiones del pie chart.
                                    fontsize=14,         #Tama침o de los textos.
                                    cmap='Pastel1')      #Para otros color maps visite: https://matplotlib.org/stable/tutorials/colors/colormaps.html
titulo = 'Distribuci칩n del n칰mero de personas que consumen alimentos altos en calorias'
plt.title(titulo)
plt.ylabel('')
plt.show()
```

```python
plt.scatter(data['FCVC'], data['NObeyesdad']);
plt.xlabel('Consume alimentos altos en calorias?');
plt.ylabel('Grado de obesidad');
plt.title('Relaci칩n entre consumo de alimentos altos en calorias y el grado de obesidad');
plt.show();
```

```python
sns.scatterplot(x=data['FCVC'].index, y=data['FCVC'],hue=data['NObeyesdad']);
plt.xlabel('Consume alimentos altos en calorias?');
plt.ylabel('Grado de obesidad');
plt.title('Relaci칩n entre consumo de alimentos altos en calorias y el grado de obesidad');
plt.show();
```

```python
fig, ax = plt.subplots()
ax.hist(data['FCVC'])
plt.show()
```

# An치lisis

Aproximadamente el 85% de los encuestas consume alimentos altos en calorias. De acuerdo al grafico de disperci칩n y de distribuci칩n, los que padecen obesidad consumen ese tipo de alimentos por encima del promedio. El histograma muestra como la mayoria de los encuestados consumen mucho este tipo de alimentos y el diagrama de caja muestra una media preocupante, as칤 como los datos atipicos parecieran ser los que consumen poco de este tipo de alimentos. Lo que concluyo entonces es que este tipo de alimentos pareciera tener un impacto directo en la obesidad y seria apropiado usarlo para modelos de machine learning.
